# Preprocess Data

```python
df_copy = df.copy()
```

```python
df_copy.isna().sum()
```

```python
len(df_copy)
```

```python
df_copy.drop(['neighbourhood_group','license'],axis=1,inplace=True)
df_copy.isna().sum()
```

```python
df_copy.dtypes
```

```python
for label, content in df_copy.items():
    if pd.api.types.is_string_dtype(content):
        print(label)
```

```python
df_copy.drop(['name','host_name'], axis=1, inplace=True)
```

```python
# Make binary columns for last_review and reviews_per_month to show whether missing values

df_copy['last_review_is_missing'] = df_copy['last_review'].isna()
df_copy['reviews_per_month_is_missing'] = df_copy['reviews_per_month'].isna()
```

```python
# Fill missing values

df_copy['last_review'].fillna('2001-01-01', inplace=True)
df_copy['reviews_per_month'].fillna(df_copy['reviews_per_month'].mean(), inplace=True)
```

```python
# Check all missing values are filled

df_copy.isna().sum()
```

### Turn last_review from object format to date format

```python
df_copy['last_review'] = pd.to_datetime(df_copy['last_review'])
df_copy['last_review'].dtype
```

### Make new columns to match the review date

```python
df_copy['last_review_year'] = df_copy['last_review'].dt.year
df_copy['last_review_month'] = df_copy['last_review'].dt.month
df_copy['last_review_ day'] = df_copy['last_review'].dt.day
```

### Drop the datetime column as not useable for model

```python
df_copy.drop('last_review',axis=1, inplace=True)
```

```python
df_copy.dtypes
```

### Convert neighbourhood and room_type objects into categorical

```python
for label, content in df_copy[['neighbourhood','room_type']].items():
    df_copy[label] = content.astype('category').cat.as_ordered()
    # Turn categories into numbers
    df_copy[label] = pd.Categorical(content).codes
```

### Convert float64 types to smaller with maximum number of decimal points

```python
df_copy['latitude'] = df_copy['latitude'].round(4).astype('float64')
df_copy['longitude'] = df_copy['longitude'].round(4).astype('float64')
df_copy['reviews_per_month'] = df_copy['reviews_per_month'].round(4).astype('float64')
```

## Modelling

```python
# Create X and y sets

X = df_copy.drop('price', axis=1)
y = df_copy['price']
```

```python
# Split into test and train sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
# Instantiate a model and fit

from sklearn.ensemble import RandomForestRegressor

regr = RandomForestRegressor()

regr.fit(X_train, y_train);
```

```python
regr.score(X_test, y_test)
```

```python
y_preds = regr.predict(X_test)
```

```python
pd.DataFrame({'Prediction':y_preds,'Actual':y_test})
```

### Finetune with RandomizedSearchCV

```python
from sklearn.model_selection import RandomizedSearchCV

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

random_search_model = RandomizedSearchCV(estimator=RandomForestRegressor(),
                                         param_distributions=param_grid,
                                         n_iter=5,
                                         cv=5,
                                         verbose=True)

random_search_model.fit(X,y);
```

```python
random_search_model.score(X_test,y_test)
```

```python
random_search_model.best_params_
```

### Create a model using the best parameters

```python
final_model = RandomForestRegressor(n_estimators=100,
                                    min_samples_split=5,
                                    min_samples_leaf=4,
                                    max_features='sqrt',
                                    max_depth=10,
                                    bootstrap=False)

final_model.fit(X_train, y_train);
```

```python
y_preds = final_model.predict(X_test)
```

```python
pd.DataFrame({'Prediction':y_preds,'Actual':y_test})
```

```python
final_model.score(X_test, y_test)
```
